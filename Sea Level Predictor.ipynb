## üåä Project Overview: Sea Level Predictor

This project is part of the **freeCodeCamp Data Analysis with Python** certification. The goal is to **analyze historical sea level data** and predict future trends using **linear regression**.

The dataset comes from the **U.S. Environmental Protection Agency (EPA)** and contains measurements of the **global average sea level** from **1880 to 2013**, adjusted based on data from the CSIRO (Commonwealth Scientific and Industrial Research Organisation).

### üîç Objectives:
- Visualize sea level rise over time using a scatter plot.
- Fit a line of best fit using **linear regression**:
  - One using the **entire dataset (1880‚Äì2013)**.
  - One using data **from the year 2000 onward**.
- Predict and extend the sea level rise trend up to the year **2050**.
- Display the two regression lines alongside the original data.

### üß∞ Tools Used:
- **Python**
- **Pandas** (for data handling)
- **Matplotlib** (for plotting)
- **SciPy's linregress** (for linear regression)

### üìÅ Dataset:
- File: `epa-sea-level.csv`
- Columns:
  - `Year`: The year of observation
  - `CSIRO Adjusted Sea Level`: Sea level in inches (adjusted)

This project demonstrates basic **time series analysis** and predictive modeling using linear regression.


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

def draw_plot():
    # Read data from file
    df = pd.read_csv("epa-sea-level.csv")
    x = df["Year"]
    y = df["CSIRO Adjusted Sea Level"]

    # Create scatter plot
    fig, ax = plt.subplots(figsize=(15, 6))
    ax.scatter(x, y, label="Original Data", color='blue', s=10)

    # Create first line of best fit (1880‚Äì2050)
    res = linregress(x, y)
    x_pred1 = pd.Series(range(1880, 2051))
    y_pred1 = res.slope * x_pred1 + res.intercept
    ax.plot(x_pred1, y_pred1, 'r', label='Best Fit Line (1880‚Äì2050)')

    # Create second line of best fit (2000‚Äì2050)
    df_recent = df[df["Year"] >= 2000]
    x_recent = df_recent["Year"]
    y_recent = df_recent["CSIRO Adjusted Sea Level"]
    res_recent = linregress(x_recent, y_recent)
    x_pred2 = pd.Series(range(2000, 2051))
    y_pred2 = res_recent.slope * x_pred2 + res_recent.intercept
    ax.plot(x_pred2, y_pred2, 'green', label='Best Fit Line (2000‚Äì2050)')

    # Add labels and title
    ax.set_xlabel("Year")
    ax.set_ylabel("Sea Level (inches)")
    ax.set_title("Rise in Sea Level")
    ax.legend()

    # Save and return plot
    plt.savefig('sea_level_plot.png')
    plt.show()
    return ax

# Call the function to display the plot
draw_plot()
```
